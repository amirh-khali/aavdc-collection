```python
import os

count = 0
# Iterate directory
for path in os.listdir('tsp/'):
    # check if current path is a file
    if os.path.isfile(os.path.join('tsp/', path)):
        count += 1
print('File count:', count)

def get_size(start_path = 'tsp/'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            # skip if it is symbolic link
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)

    return total_size

print("Size:", get_size()//1e6, 'MB')
```

```python
import os

videos_to_process = {'video_path': [], 'feature_path': []}

dir = base_dir + '104/'
feat_dir = base_dir + '104/feats/'
for path in os.listdir(dir):
    if os.path.isfile(os.path.join(dir, path)):
        videos_to_process['video_path'].append(os.path.join(dir, path))
        videos_to_process['feature_path'].append(os.path.join(feat_dir, path).replace('.mp4', '.npy').replace('.mkv', '.npy').replace('.webm', '.npy'))
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(videos_to_process)
df.to_csv('input.csv')
```
